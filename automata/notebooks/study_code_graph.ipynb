```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns

from automata.configs.config_enums import ConfigCategory
from automata.tools.search.symbol_converter import SymbolConverter
from automata.tools.search.symbol_graph import SymbolGraph
from automata.tools.search.symbol_rank.symbol_embedding_map import SymbolEmbeddingMap
from automata.tools.search.symbol_rank.symbol_similarity import SymbolSimilarity
```

```python
file_dir = os.getcwd()

scip_path = os.path.join(
    file_dir, "..", "configs", ConfigCategory.SYMBOLS.value, "index.scip"
)
embedding_path = os.path.join(
    file_dir, "..", "configs", ConfigCategory.SYMBOLS.value, "symbol_embedding.json"
)
```

```python
symbol_converter = SymbolConverter()
symbol_graph = SymbolGraph(scip_path, symbol_converter)
symbol_subgraph = symbol_graph.symbol_to_symbol_subgraph
```

```python
import networkx as nx
from pyvis.network import Network

# Create a copy of the graph to avoid modifying the original.
symbol_subgraph_copy = symbol_subgraph.copy()

# Relabel all nodes to strings.
mapping = {node: str(node.uri).split("/")[1] for node in symbol_subgraph_copy.nodes()}
symbol_subgraph_string_nodes = nx.relabel_nodes(symbol_subgraph_copy, mapping)

node = "AutomataAgentConfigBuilder#"
```

```python
ego_graph = nx.ego_graph(symbol_subgraph_string_nodes, node)

# Then, try creating the PyVis network again.
nt = Network('500px', '800px', notebook=True)
nt.from_nx(ego_graph)
nt.toggle_physics(False)  # Disable physics simulation
nt.show('nx.html')
```
